```python
# Custom Exception
class LoginError(Exception):
    def __init__(self, message):
        super().__init__(message)

# Correct credentials
correct_username = "admin"
correct_password = "1234"

attempts = 3

while attempts > 0:
    try:
        username = input("Enter username: ")
        password = input("Enter password: ")

        if username == correct_username and password == correct_password:
            print("Login Successful!")
            break
        else:
            attempts -= 1
            print("Invalid credentials!")

            if attempts == 0:
                raise LoginError("Too many failed attempts. Access Denied!")
            else:
                print("Attempts left:", attempts)

    except LoginError as e:
        print(e)
```

```python
# Custom Exception
class InsufficientBalanceError(Exception):
    pass

balance = 5000  # initial balance

try:
    amount = float(input("Enter withdrawal amount: "))

    if amount > balance:
        raise InsufficientBalanceError("Insufficient balance!")

    elif amount <= 0:
        print("Enter a valid amount!")

    else:
        balance -= amount
        print("Withdrawal successful!")
        print("Remaining balance:", balance)

except InsufficientBalanceError as e:
    print("Error:", e)

except ValueError:
    print("Error: Please enter a valid number!")

except Exception as e:
    print("Unexpected error:", e)
```

```python
# Custom Exceptions
class InvalidAgeError(Exception):
    pass

class UnderAgeError(Exception):
    pass

try:
    age = int(input("Enter your age: "))

    if age <= 0:
        raise InvalidAgeError("Age must be a positive number!")

    if age < 18:
        raise UnderAgeError("Age must be 18 or above!")

    print("Valid age. Access granted!")

except ValueError:
    print("Error: Please enter a valid integer!")

except InvalidAgeError as e:
    print("Error:", e)

except UnderAgeError as e:
    print("Error:", e)

except Exception as e:
    print("Unexpected error:", e)
```

```python
# Smart File-Based User Management System (Single File)

from datetime import datetime

# Custom Exceptions
class UserExistsError(Exception):
    pass

class InvalidLoginError(Exception):
    pass


# File Handling Functions
def read_users():
    try:
        with open("users.txt", "r") as f:
            return f.readlines()
    except FileNotFoundError:
        return []

def write_user(user):
    with open("users.txt", "a") as f:
        f.write(user + "\n")


# Log Function
def log_activity(text):
    with open("log.txt", "a") as f:
        f.write(f"{datetime.now()} - {text}\n")


# Register Function
def register():
    username = input("Enter username: ")
    password = input("Enter password: ")

    users = read_users()

    for user in users:
        if username == user.split(",")[0]:
            raise UserExistsError("⚠ User already exists!")

    write_user(username + "," + password)
    print(" Registered successfully!")
    log_activity(f"{username} registered")


# Login Function
def login():
    attempts = 3

    while attempts > 0:
        username = input("Enter username: ")
        password = input("Enter password: ")

        users = read_users()

        for user in users:
            u, p = user.strip().split(",")
            if u == username and p == password:
                print("Login successful!")
                log_activity(f"{username} logged in")
                dashboard(username)
                return

        attempts -= 1
        print(f"Invalid credentials! Attempts left: {attempts}")

    raise InvalidLoginError("Too many failed attempts!")


# Dashboard
def dashboard(username):
    while True:
        print("\n--- DASHBOARD ---")
        print("1. View Profile")
        print("2. Delete Account")
        print("3. Logout")

        choice = input("Enter choice: ")

        if choice == "1":
            print(f" Username: {username}")

        elif choice == "2":
            delete_user(username)
            print(" Account deleted!")
            log_activity(f"{username} deleted account")
            break

        elif choice == "3":
            print("Logged out!")
            log_activity(f"{username} logged out")
            break

        else:
            print("Invalid choice!")


# Delete User
def delete_user(username):
    users = read_users()

    with open("users.txt", "w") as f:
        for user in users:
            if not user.startswith(username + ","):
                f.write(user)


# Main Menu
def main():
    while True:
        print("\n===== MENU =====")
        print("1. Register")
        print("2. Login")
        print("3. Exit")

        choice = input("Enter choice: ")

        try:
            if choice == "1":
                register()

            elif choice == "2":
                login()

            elif choice == "3":
                print("Exiting program...")
                break

            else:
                print("Invalid choice!")

        except Exception as e:
            print("Error:", e)


#  Run Program
if __name__ == "__main__":
    main()
```
